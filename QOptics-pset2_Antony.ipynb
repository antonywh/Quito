# Qn 1

```python
####################################################################################
### Qn 1 - Sorry for the inelegant solution! -
####################################################################################
# Initialize the states
N = 20 # Levels in the Hilbert space

a = destroy(N)
# n_oper_3 = tensor(a.dag()*a,qeye(N))
# n_oper_4 = tensor(qeye(N),a.dag()*a)
n_oper = a.dag()*a

psig = fock(N,0)
psi = tensor(psig,psig)

alpha = 2**0.5
alpha_neg = -alpha
s_a = 0.5*np.exp(1j*0)
s_b = 0.5*np.exp(1j*0)
phi = np.pi
s_bb = 0.5*np.exp(1j*phi)

################ Qn1_1 ################
c_3 = tensor(a.dag(),qeye(N))
c_4 = tensor(qeye(N),a.dag())
c_1 = (1/(2**0.5))*(c_3+1j*c_4)
c_2 = (1/(2**0.5))*(c_3-1j*c_4)

# c_2*psi == (1/(2**0.5))*(tensor(fock(N,1),fock(N,0))-tensor(fock(N,0),fock(N,1)))

qn1_1 = (c_1*c_2*psi).unit()
qn1_1_3 = qn1_1.ptrace(0)
qn1_1_4 = qn1_1.ptrace(1)

################ Qn1_2 ################
disp_3 = tensor(((1/2**0.5)*(alpha*a.dag()-alpha*a)).expm(),qeye(N))
disp_4 = tensor(qeye(N),(1j*(1/2**0.5)*(alpha*a.dag()-alpha*a)).expm())
disp_1 = (disp_3*disp_4)

# tensor(coherent(N,alpha/2**0.5),coherent(N,alpha/2**0.5)) == disp_a*psi
qn1_2 = (disp_1*psi).unit()
qn1_2_3 = qn1_2.ptrace(0)
qn1_2_4 = qn1_2.ptrace(1)

################ Qn1_3 ################
disp_3_neg = tensor(((1/2**0.5)*(alpha*a.dag()-alpha*a)).expm().dag(),qeye(N))
disp_4_neg = tensor(qeye(N),((1/2**0.5)*(alpha*a.dag()-alpha*a)).expm().dag())
disp_1_neg = (disp_3_neg*disp_4_neg)

# disp_a*psi + disp_a_neg*psi == tensor(coherent(N,alpha/2**0.5),coherent(N,alpha/2**0.5)) + tensor(coherent(N,-alpha/2**0.5),coherent(N,-alpha/2**0.5))
qn1_3 = (disp_1*psi + disp_1_neg*psi).unit()
qn1_3_3 = qn1_3.ptrace(0)
qn1_3_4 = qn1_3.ptrace(1)

################ Qn1_4a ################
sqiz_c = tensor(((1/4)*((s_a.conj()+s_b.conj())*(a**2)-
                        (s_a+s_b)*a.dag()**2)+
                         (2*a*(s_a.conj()-s_b.conj()))+
                         (2*a.dag()*(s_b-s_a))).expm(),qeye(N))
sqiz_d = tensor(qeye(N),((-1/4)*((s_a.conj()+s_b.conj())*(a**2)-
                        (s_a+s_b)*a.dag()**2)+
                         (2j*a*(s_a.conj()-s_b.conj()))+
                         (2j*a.dag()*(s_b-s_a))).expm())


qn1_4a = (sqiz_c*sqiz_d*psi).unit()
qn1_4a_3 = qn1_4a.ptrace(0)
qn1_4a_4 = qn1_4a.ptrace(1)


################ Qn1_4b ################
sqiz_cb = tensor(((1/4)*((s_a.conj()+s_bb.conj())*(a**2)-
                        (s_a+s_bb)*a.dag()**2)+
                         (2*a*(s_a.conj()-s_bb.conj()))+
                         (2*a.dag()*(s_bb-s_a))).expm(),qeye(N))
sqiz_db = tensor(qeye(N),((-1/4)*((s_a.conj()+s_bb.conj())*(a**2)-
                        (s_a+s_bb)*a.dag()**2)+
                         (2j*a*(s_a.conj()-s_bb.conj()))+
                         (2j*a.dag()*(s_bb-s_a))).expm())


qn1_4b = (sqiz_cb*sqiz_db*psi).unit()
qn1_4b_3 = qn1_4b.ptrace(0)
qn1_4b_4 = qn1_4b.ptrace(1)


# Initalize plot variables
xvec = np.linspace(-5,5,200)
yvec = xvec

# Calculate the wigner function for the states - 3 is for output port 1, 4 is for output port 2
W_qn1_1_3 = wigner(qn1_1_3, xvec, yvec)
W_qn1_1_4 = wigner(qn1_1_4, xvec, yvec)
W_qn1_2_3 = wigner(qn1_2_3, xvec, yvec)
W_qn1_2_4 = wigner(qn1_2_4, xvec, yvec)
W_qn1_3_3 = wigner(qn1_3_3, xvec, yvec)
W_qn1_3_4 = wigner(qn1_3_4, xvec, yvec)
W_qn1_4a_3 = wigner(qn1_4a_3, xvec, yvec)
W_qn1_4a_4 = wigner(qn1_4a_4, xvec, yvec)
W_qn1_4b_3 = wigner(qn1_4b_3, xvec, yvec)
W_qn1_4b_4 = wigner(qn1_4b_4, xvec, yvec)

# create a plot for 3 figures
fig, axes = plt.subplots(5, 2, figsize=(9,20))
axes = axes.flatten()

# create contour plots using the results of the wigner func
# The colour distribution varies from that in the lecture notes as it is not normalized to -1 to 1
cont0 = axes[0].contourf(xvec, xvec, W_qn1_1_3, 100, cmap = 'seismic')
cont1 = axes[1].contourf(xvec, xvec, W_qn1_1_4, 100, cmap = 'seismic')
cont2 = axes[2].contourf(xvec, xvec, W_qn1_2_3, 100, cmap = 'seismic')
cont3 = axes[3].contourf(xvec, xvec, W_qn1_2_4, 100, cmap = 'seismic')
cont4 = axes[4].contourf(xvec, xvec, W_qn1_3_3, 100, cmap = 'seismic')
cont5 = axes[5].contourf(xvec, xvec, W_qn1_3_4, 100, cmap = 'seismic')
cont6 = axes[6].contourf(xvec, xvec, W_qn1_4a_3, 100, cmap = 'seismic')
cont7 = axes[7].contourf(xvec, xvec, W_qn1_4a_4, 100, cmap = 'seismic')
cont8 = axes[8].contourf(xvec, xvec, W_qn1_4b_3, 100, cmap = 'seismic')
cont9 = axes[9].contourf(xvec, xvec, W_qn1_4b_4, 100, cmap = 'seismic')

# add colorbar
cb0 = fig.colorbar(cont0,ax = axes[0])
cb1 = fig.colorbar(cont1,ax = axes[1])
cb2 = fig.colorbar(cont2,ax = axes[2])
cb3 = fig.colorbar(cont3,ax = axes[3])
cb4 = fig.colorbar(cont4,ax = axes[4])
cb5 = fig.colorbar(cont5,ax = axes[5])
cb6 = fig.colorbar(cont6,ax = axes[6])
cb7 = fig.colorbar(cont7,ax = axes[7])
cb8 = fig.colorbar(cont8,ax = axes[8])
cb9 = fig.colorbar(cont9,ax = axes[9])

# set the title for each plot
lbl0 = axes[0].set_title("qn1_1 output port 1")
lbl1 = axes[1].set_title("qn1_1 output port 2")
lbl2 = axes[2].set_title("qn1_2 output port 1")
lbl3 = axes[3].set_title("qn1_2 output port 2")
lbl4 = axes[4].set_title("qn1_3 output port 1")
lbl5 = axes[5].set_title("qn1_3 output port 2")
lbl6 = axes[6].set_title("qn1_4a output port 1")
lbl7 = axes[7].set_title("qn1_4a output port 2")
lbl8 = axes[8].set_title("qn1_4b ($\phi$ = $\pi$) output port 1")
lbl9 = axes[9].set_title("qn1_4b ($\phi$ = $\pi$) output port 2")

#add a text below plots
axes[0].text(4,0.4, '$n_3$ = {:.2f} \n$n_4$ = {:.2f} \nEigenenergy of one of the state = {:.2f} \nentangled = Yes'.format(expect(n_oper,qn1_1_3),expect(n_oper,qn1_1_4),qn1_1.ptrace(0).eigenstates(sort='high')[0][0]), 
             size=12, ha="center", transform=axes[0].transAxes);
axes[2].text(4,-0.8, '$n_3$ = {:.2f} \n$n_4$ = {:.2f} \nEigenenergy of one of the state = {:.2f} \nentangled = No'.format(expect(n_oper,qn1_2_3),expect(n_oper,qn1_2_4),qn1_2.ptrace(0).eigenstates(sort='high')[0][0]), 
             size=12, ha="center", transform=axes[0].transAxes);
axes[4].text(4,-2, '$n_3$ = {:.2f} \n$n_4$ = {:.2f} \nEigenenergy of one of the state = {:.2f} \nentangled = Yes'.format(expect(n_oper,qn1_3_3),expect(n_oper,qn1_3_4),qn1_3.ptrace(0).eigenstates(sort='high')[0][0]), 
             size=12, ha="center", transform=axes[0].transAxes);
axes[6].text(4,-3.2, '$n_3$ = {:.2f} \n$n_4$ = {:.2f} \nEigenenergy of one of the state = {:.2f} \nentangled = No'.format(expect(n_oper,qn1_4a_3),expect(n_oper,qn1_4a_4),qn1_4a.ptrace(0).eigenstates(sort='high')[0][0]), 
             size=12, ha="center", transform=axes[0].transAxes);
axes[8].text(4,-4.4, '$n_3$ = {:.2f} \n$n_4$ = {:.2f} \nEigenenergy of one of the state = {:.2f} \nentangled = No'.format(expect(n_oper,qn1_4b_3),expect(n_oper,qn1_4b_4),qn1_4b.ptrace(0).eigenstates(sort='high')[0][0]), 
             size=12, ha="center", transform=axes[0].transAxes);

###### the way to determine entangled states was learnt from Prof. Velario through Peter Sidajaya ######
```

# Qn 2

```python
N = 20

#cavity mode |0>
cav = fock(N,0)

#two level system (TLS), set ground state as |0>
g = fock(2,0)
cav_g = tensor(cav,g)

#applying operator to bring TLS g to equal superposition of g+e
H = hadamard_transform(1) # we will talk about this
HI = tensor(qeye(N),H)
cav_ge = HI*cav_g

#applying displacement operator with alpha = 2
a = destroy(N)
alpha = 2
disp = ((alpha*a.dag()-alpha*a)).expm()
disp_1 = tensor(disp,qeye(2))
disp_cavge = disp_1*cav_ge

#evolving the system with Hamiltonian H for time T = pi/chi12
w1 = 100 # cavity frequency= w1
w2 = 110 # TLS frenqency= w2
chi12 = 10 # coupling strength= chi12

e = tensor(qeye(N),fock(2,1))

# cavity mode operator
ac = tensor(destroy(N), qeye(2))


# Hamiltonian
H = w1*ac.dag()*ac + w2*e*e.dag() - chi12*ac.dag()*ac*e*e.dag()
psi = disp_cavge

# list of times for which the solver should store the state vector
tlist = np.linspace(0, np.pi/chi12, 100)
final = mesolve(H, psi, tlist, [], [])
# final.states[-1]

#applying displacement operator with alpha = 2
disp_2 = tensor(disp, qeye(2))
disp_2nd = disp_2*final.states[-1]

#Applying rotation on TLS when cav is in vaccuum and nothing when cav is not in vaccuum
rot = tensor(ket2dm(cav),sigmax()) + tensor((qeye(N)-ket2dm(cav)),qeye(2))
rotated = rot*disp_2nd

#Applying displacement operator with alpha = -2
disp_negb = disp.dag() #because D(-alpha) = D(alpha).dag()
disp_f = tensor(disp_negb,qeye(2))
disp_final = disp_f*rotated

#The final cavity state
f_cav = disp_final.ptrace(0).eigenstates(sort='high')[1][0]

#The final TLS state
f_tls = disp_final.ptrace(1).eigenstates(sort='high')[1][0]

#to a global phase rotation to make it look nicer
ff_tls = globalphase(np.pi)*f_tls


# print('The final cavity state is {} \nThe final TLS state is {}'.format(f_cav,ff_tls))


##### Plotting the wigner #####
# Initalize plot variables
xvec = np.linspace(-5,5,200)
yvec = xvec

# Calculate the wigner function for the states - 3 is for output port 1, 4 is for output port 2
W_tls = wigner(disp_final.ptrace(1).eigenstates(sort='high')[1][0], xvec, yvec)
W_cav= wigner(disp_final.ptrace(0).eigenstates(sort='high')[1][0], xvec, yvec)
W_cat= wigner(coherent(N,2)-coherent(N,-2), xvec, yvec)

# create a plot for 3 figures
fig, axes = plt.subplots(1, 3, figsize=(24,6))

# create contour plots using the results of the wigner func
# The colour distribution varies from that in the lecture notes as it is not normalized to -1 to 1
cont0 = axes[0].contourf(xvec, xvec, W_tls, 100, cmap = 'seismic')
cont1 = axes[1].contourf(xvec, xvec, W_cav, 100, cmap = 'seismic')
cont2 = axes[2].contourf(xvec, xvec, W_cat, 100, cmap = 'seismic')

# add colorbar
cb0 = fig.colorbar(cont0,ax = axes[0])
cb1 = fig.colorbar(cont1,ax = axes[1])
cb2 = fig.colorbar(cont2,ax = axes[2])

# set the title for each plot
lbl0 = axes[0].set_title("TLS state")
lbl1 = axes[1].set_title("Cavity state")
lbl2 = axes[2].set_title("Two-component cat state for reference")
```
